```python
# lets find the class distribution in target
wine_df.quality.value_counts()
```

```python
# lets perform one hot encoding on target column
wine_df1=pd.get_dummies(data=wine_df,columns=['quality'])
```

```python
wine_df1.head()
```

```python
wine_df1.shape
```

```python
# identify X and y
X=wine_df1.drop(['quality_3', 'quality_4', 'quality_5',
       'quality_6', 'quality_7', 'quality_8'],axis=1)
X.head()
```

```python
y=wine_df1[wine_df1.columns[-6:]]
y.head()
```

```python
# lets perform train test split
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.3,random_state=21)
```

```python
print(X_train.shape)
print(X_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
# lets apply standard scaler
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
sc.fit(X_train)
X_train=sc.transform(X_train)
X_test=sc.transform(X_test)
```

```python
print(X_train.shape)
print(X_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
import tensorflow as tf
from tensorflow import keras
```

```python
#lets create neural network
model=keras.Sequential([
                        keras.layers.Dense(11,input_shape=(11,),activation='relu'),
                        keras.layers.Dense(10,activation='relu'),
                        keras.layers.Dense(6,activation='softmax')
])
```

```python
model.compile(loss='categorical_crossentropy',optimizer='adam',metrics=['accuracy'])
```

```python
from keras.callbacks import EarlyStopping
from keras.callbacks import ModelCheckpoint
from keras.models import load_model
# lets define early stopping and model checkpoint
es=EarlyStopping(monitor='val_loss',mode='min',verbose=1,patience=200)
mc=ModelCheckpoint('best_model.h5',monitor='val_accuracy',mode='max',verbose=1,save_best_only=True)
#fit the model
history=model.fit(X_train,y_train,validation_data=(X_test,y_test),epochs=4000,verbose=0,callbacks=[es,mc])
# load the saved model
saved_model=load_model('best_model.h5')
# evaluate model
_,train_acc=saved_model.evaluate(X_train,y_train,verbose=0)
_,test_acc=saved_model.evaluate(X_test,y_test,verbose=0)
print('Train: {}, Test: {}'.format(train_acc,test_acc))
```

```python
#lets plot the train and test loss
plt.plot(history.history['loss'],label='train')
plt.plot(history.history['val_loss'],label='test')
plt.legend()
plt.show()
```
